### Avaliação dos melhores classificadores

```python
import pandas as pd
import scipy
import numpy as np
import seaborn as sns
# warn
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
# cross validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
# Algoritmos
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
# Friedman test
from scipy.stats import friedmanchisquare
```

### Read data

```python
X_zscore = pd.read_csv("./data_pre/X_zscore.data", index_col=0)
X = pd.read_csv("./data_pre/X.data", index_col=0)
y = pd.read_csv("./data_pre/y.data", index_col=0, header=None)
```

### Split training data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

X_train_zscore, X_test_zscore, y_train, y_test = train_test_split(
    X_zscore, y, test_size=0.3, random_state=42)
```

### Classificadores

```python
means = {'tree': [], 'mlp': [], 'knn': [], 'rf': []}
rs = []

for i in range(30):
    print("Conjunto: ", i)
    print(means)
    # clfs
    tree = DecisionTreeClassifier(criterion='entropy', splitter='best', presort=True)
    mlp = MLPClassifier(activation='relu', solver='adam', early_stopping=True, 
                        max_iter=1000, batch_size=16, validation_fraction=0.2,
                        hidden_layer_sizes=(100, 50))
    knn = KNeighborsClassifier(n_neighbors=5, weights='distance', metric='minkowski',
                               p=1)
    rf = RandomForestClassifier(n_estimators=2000 , criterion='gini', 
                                random_state=42)
    # kfold
    random_state = np.random.randint(200)
    rs.append(random_state)
    kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=random_state)
    
    # scores
    print('Tree')
    means['tree'].append(cross_val_score(tree, X_train, y_train, cv=kfold, n_jobs=-1).mean())
    #print('RF')
    #means['rf'].append(cross_val_score(rf, X_train, y_train, cv=kfold).mean())
    #print('MLP')
    means['mlp'].append(cross_val_score(mlp, X_train_zscore, y_train, cv=kfold, n_jobs=2).mean())
    print('KNN')
    means['knn'].append(cross_val_score(knn, X_train_zscore, y_train, cv=kfold, n_jobs=2).mean())
```

```python
rf_mean = []
for i in range(30):
    print("Iter: ", i)
    rf = RandomForestClassifier(n_estimators=2000 , criterion='gini', 
                                random_state=42)
    kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=rs[i])

    score = cross_val_score(rf, X_train, y_train, cv=kfold, n_jobs=2).mean()
    rf_mean.append(score)
    print(rf_mean)
```
